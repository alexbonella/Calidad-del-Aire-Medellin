```python
estacion=[]
fecha=[]
calidad=[]
valor=[]
for estaciones in range(len(df_data['codigoSerial'])):
    print('aqui vamos por la estacion : ' + str(estaciones))
    for data in range(len(df_data['datos'][estaciones])):
        estacion.append(df_data['codigoSerial'][estaciones])
        fecha.append(df_data['datos'][estaciones][data]['fecha'])
        calidad.append(df_data['datos'][estaciones][data]['calidad'])
        valor.append(df_data['datos'][estaciones][data]['valor'])        
```

```python
datos={'Estacion':estacion,'Fecha':fecha,'Valor PM25':valor,'Calidad':calidad}
```

```python
df_final=pd.DataFrame(datos)
```

```python
estacion=[]
fecha=[]
calidad=[]
valor=[]
variable=[]
for estaciones in range(len(datos_ozono['codigoSerial'])):
    print('aqui vamos por la estacion : ' + str(estaciones))
    for data in range(len(datos_ozono['datos'][estaciones])):
        estacion.append(datos_ozono['codigoSerial'][estaciones])
        fecha.append(datos_ozono['datos'][estaciones][data]['fecha'])
        calidad.append(datos_ozono['datos'][estaciones][data]['calidad'])
        valor.append(datos_ozono['datos'][estaciones][data]['valor'])
        variable.append(datos_ozono['datos'][estaciones][data]['variableConsulta'])   
```

```python
datos={'Estacion':estacion,'Fecha':fecha,'Valor Ozono':valor,'Calidad':calidad,'Variable Consulta':variable}
```
