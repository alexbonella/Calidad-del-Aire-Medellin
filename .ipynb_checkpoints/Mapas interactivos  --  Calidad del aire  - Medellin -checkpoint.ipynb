# Funciones creacion dinamica de mapas

Las funciones utilizadas fueron basadas en el siguiente post : 

https://blog.prototypr.io/interactive-maps-in-python-part-3-29f14a9b2f7d

```python
def plot_map(df_medicion, colored=False, zoom_start=10):
    
    folium_map = folium.Map(location=[lon_center, lat_center],
                            tiles='cartodb positron', zoom_start=zoom_start)

    for i, row in df_medicion.iterrows():
        if colored:
            
            folium.CircleMarker(location=[row.latitud, row.longitud], radius=row.radius, color=row.color,
                                fill=True, fill_opacity=0.4).add_to(folium_map)
            folium_map.add_child(colormap)    
            
        else:
            folium.CircleMarker(location=[row.latitud, row.longitud], radius=row.radius, color="#0A8A9F",
                                fill=True, fill_opacity=0.4).add_to(folium_map)


    return folium_map
```

```python
def make_hourly_interpolations_df(estacion_data, hora):
    
    hora_min = int(hora)
    hora_max = int(hora)+1
    
    # Manejamos cambio de día de las 23 a las 0 horas por separado
    if hora_max==24:
        hora_min = 23
        hora_max = 0
        
    data_hora = estacion_data[estacion_data.hora.isin([hora_min, hora_max])]
    
    x0 = data_hora[data_hora.hora==hora_min].Valor.values
    x1 = data_hora[data_hora.hora==hora_max].Valor.values
    
    interpolated_df = [data_hora[data_hora.hora==hora_min]]
    
    for h in [1+i*.2 for i in range(1, 5)]:   
        try:
            new_row = data_hora.copy()[0:1]
            new_row['hora'] = hora+(h%1)
            new_row['Valor'] = x0*(1-h%1)+x1*(h%1)

            interpolated_df.append(new_row)
            
        except ValueError:
            display(data_hora)
            
    interpolated_df.append(data_hora[data_hora.hora==hora_max])
        
    return pd.concat(interpolated_df, ignore_index=True)

def make_daily_interpolations_df(data_estacion):
    interpolated_data = []
    for h in range(0, 24):
        aux = make_hourly_interpolations_df(data_estacion, h)
        interpolated_data.append(aux)
    
    interpolated_data = pd.concat(interpolated_data, ignore_index=True).drop_duplicates()
    
    return interpolated_data
```

```python
import warnings
warnings.filterwarnings("ignore")

from time import sleep
from shapely.geometry import Point, Polygon, MultiPoint
from selenium import webdriver

route="/home/alexbonella2806/Geolocalizacion/map_frames/geckodriver"
save_path=os.getcwd()+'/map_frames/'

def go_ozone_frame(i, hour_of_day, data, save_path, title, zoom):
    
    
    
    # create the map object
    hour_data = data[data.hora==hour_of_day]
    my_frame = plot_map(hour_data, colored=True, zoom_start=zoom)
    my_frame.save(os.getcwd()+'/map_frames/test.html')
    
    browser = webdriver.Firefox(executable_path=route)
    fn='test.html'
    tmpurl='file://{path}/{mapfile}'.format(path=os.getcwd()+'/map_frames/',mapfile=fn)
    my_frame.save(fn)

    browser.get(tmpurl)

    time.sleep(1.5)
    browser.save_screenshot(os.getcwd()+'/map_frames/map.png')
    with open(os.getcwd()+'/map_frames/map.png', "rb") as image:
        f = image.read()
        png = bytearray(f)
    browser.quit()

    
    hour = int(hour_of_day)
    minutes = round((hour_of_day % 1) * 60)
    
    # create a PIL image object
    image = Image.open(io.BytesIO(png))
    draw = ImageDraw.ImageDraw(image)
    
    # load a font
    font = ImageFont.truetype('/home/alexbonella2806/Geolocalizacion/symbola/Symbola_hint.ttf', size=30)
    
     # draw time of day text
    draw.text((20,image.height - 50), 
              "Hora: {:0>2}:{:0>2}".format(hour, minutes),
              fill=(0, 0, 0), 
              font=font)
    
    # draw title
    draw.text((100, 20), 
              title,
              fill=(0, 0, 0), 
              font=font)
    
    # write to a png file
    filename = os.path.join(save_path, "frame_{:0>5}.png".format(i))
    image.save(filename, "PNG")
    return
```

```python
o3_hora=df_geo_final
```

```python
sc_o3_hora = o3_hora.copy()
sc_o3_hora['radius'] = np.sqrt((12*(sc_o3_hora.Valor - 
                        sc_o3_hora.Valor.min())/(sc_o3_hora.Valor.max() - 
                                                  sc_o3_hora.Valor.min()) + 1)**3)

o3_hora = o3_hora.groupby('Estacion').apply(make_daily_interpolations_df).reset_index(drop=True)

o3_hora['radius'] = np.sqrt((12*(o3_hora.Valor - 
                        o3_hora.Valor.min())/(o3_hora.Valor.max() - 
                                                  o3_hora.Valor.min()) + 1)**3)

colormap = cm.LinearColormap(list(map(mcolors.to_hex, mtplb_cm.inferno.colors)),
                             vmin=o3_hora.Valor.min(),
                             vmax=o3_hora.Valor.max()).to_step(10)

colormap.caption = 'Valores promedio PM2.5 por Hora'

o3_hora['color'] = o3_hora.Valor.map(colormap)

# Vamos a usar esto para centrar el mapa
aux = o3_hora[['latitud', 'longitud']].values
centroid = MultiPoint(aux).convex_hull.centroid
lat_center, lon_center = centroid.y, centroid.x

print(o3_hora.shape, o3_hora.Estacion.nunique())
o3_hora.head()
```
